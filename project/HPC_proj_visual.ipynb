### Quick sort 

```python
# time evaluation
import numpy as np
import matplotlib.pyplot as plt
```

```python
%time !/usr/local/Cellar/open-mpi/4.0.3/bin/mpicc -o qsort qsort.c -lm
```

```python
for i in range(1, 5):
    
    !/usr/local/Cellar/open-mpi/4.0.3/bin/mpirun --use-hwthread-cpus --oversubscribe -np $i qsort input100.txt out.txt
```

```python
#time dependancy on num of threads:
communication_time = np.array([0.000070,0.000410, 0.000218, 0.000277, 0.002069])
plt.title("Time avaluation on 100 data len")
plt.plot(np.arange(1, 6), communication_time, label = "Quick sort")
plt.legend()
plt.xlabel("Num of processors")
plt.ylabel("Time elapsed")
plt.show()

```

### Radix sort

```python
# radix sort time evaluation
%time !/usr/local/Cellar/open-mpi/4.0.3/bin/mpicc -o radix radixs/radixbalanced.c -lm
```

```python
for i in range(1, 6):
    
    !/usr/local/Cellar/open-mpi/4.0.3/bin/mpirun --use-hwthread-cpus --oversubscribe -np $i radix
```

```python
communication_time = np.array([0.000191,0.000362, 0.000785, 0.002766, 0.004420])
plt.title("Time avaluation on 128 data len")
plt.plot(np.arange(1, 6), communication_time, label = "Radix sort")
plt.legend()
plt.xlabel("Num of processors")
plt.ylabel("Time elapsed")
plt.show()
```

```python
#time dependancy on num of threads:
communication_time_rd = np.array([1.002094,0.626361, 0.568873, 0.761646,0.699093])
plt.title("Time avaluation on 2^20 data len")
plt.plot(np.arange(1, 6), communication_time_rd, label = "Radix sort")
plt.legend()
plt.xlabel("Num of processors")
plt.ylabel("Time elapsed")
plt.show()
```

### Methods comparison:

```python
plt.figure(figsize=(12, 11))
plt.title("Comparison of sorting algorithms", fontsize=20)
plt.plot(np.arange(1, 6), communication_time_shear, label = "Shear")
plt.plot(np.arange(1, 6), communication_time_rd, label = "Radix")
plt.plot(np.arange(1, 6), communication_time_qs, label = "QuickSort")
plt.xlabel("Num of processors", fontsize=18)
plt.ylabel("Time elapsed", fontsize=18)
plt.grid()
plt.legend()
plt.show()
```
